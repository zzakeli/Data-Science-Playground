```python
"""
Imports tools and libraries used 
for dashboard summary and analytics
"""

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/college_student_placement_dataset.csv')
print(df.head())
```

```python
"""
Checks for missing values
"""
print(df.isna().any())
```

```python
"""
Visualization of Student's Placement
"""
plt.hist(df['Placement'],bins=2)
plt.title("College Student's Placement")
plt.xlabel("Placement")
plt.ylabel("Number of Students")
plt.show()
```

```python
"""
Number of Student's Placement
"""
print(df.groupby('Placement')['Placement'].value_counts())
```

```python
"""
Feature Engineering
Creates Soft Skills using Extra Curricular Score and Communication Skills
"""
df['Soft_Skills'] = (df['Extra_Curricular_Score'] + df['Communication_Skills']) / 2
# print(df[['College_ID', 'Soft_Skills']].sort_values(by='Soft_Skills', ascending=False))
print(df[['College_ID', 'Soft_Skills']].head())
```

```python
"""
Visualization of College Student's IQ
"""
# df.groupby('IQ')['IQ'].value_counts().plot(kind='bar', title="College Student's IQ")
plt.hist(df['IQ'], bins=20)
plt.ylabel("Number of College Students")
plt.xlabel("IQ")
plt.title("College Student's IQ")
plt.show()
```

```python
"""
Visualization of College Internship Experience
"""

df.groupby('Internship_Experience')['Internship_Experience'].value_counts().plot(xlabel='Internship Experience', ylabel='Number of Students',kind='bar', title='College Internship Experience')
plt.show()
```

```python
"""
Number of students who has Internship Experience
"""
print(df.groupby('Internship_Experience')['Internship_Experience'].value_counts()) 
```

```python
"""
Average IQ, Previous Sem Result, CGPA
"""
print(df.loc[:, ['IQ', 'Prev_Sem_Result', 'CGPA']].mean())
```

```python
"""
A student with highest IQ
"""
print(df.loc[df['IQ'] == df['IQ'].max()][['College_ID', 'IQ']])
```

```python
"""
Students with highest Previous Sem Result
"""
print(df.loc[df['Prev_Sem_Result'] == df['Prev_Sem_Result'].max()][['College_ID', 'Prev_Sem_Result']])
```

```python
"""
Students that excels on Academics
"""
print(df.loc[df['Academic_Performance'] == df['Academic_Performance'].max()][['College_ID','Academic_Performance']])
```
